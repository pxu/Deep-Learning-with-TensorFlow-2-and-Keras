```python
import collections

```

#Simulating a small dictionary of words and their countsIn real project, will count words from a large text file.


```python
word_counts = collections.Counter([
    "the", "the", "the", "longest", "longest", "list", "of", "stuff", "at", "domain"
])
TOTAL_WORDS = sum(word_counts.values())
```

#Calculates the probability of a single word.
    


```python
def prob_word(word):
    if word in word_counts:
        return word_counts[word] / TOTAL_WORDS
    else:
        # For unknown words: make it less likely the longer the word is
        return 1 / (TOTAL_WORDS * (10 ** len(word)))
```

#Segmentation of words without spaces


```python
def segment(text):
    length = len(text)
    max_prob = [0] * (length + 1)
    max_prob[length] = 1
    split_idx = [-1] * (length + 1)

    # Loop backward through the string
    for start in range(length - 1, -1, -1):
        for split in range(start + 1, length + 1):

            # Extract the current word chunk
            word_chunk = text[start:split]

            # Calculate the probability
            p = max_prob[split] * prob_word(word_chunk)

            # If this split is better, save it
            if p > max_prob[start]:
                max_prob[start] = p
                split_idx[start] = split

    # Reconstruct the words using the split indices
    i = 0
    words = []
    while i < length:
        next_split = split_idx[i]
        if next_split == -1:
            return None # Failed to segment
        words.append(text[i:next_split])
        i = next_split

    return words
```

### Quick Test


```python
test_string = "thelongestlistofthelongeststuffatthelongestdomainnameatlonglast.com"
print("Result:", segment(test_string))
```
